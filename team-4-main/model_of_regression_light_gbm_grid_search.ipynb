```python
from sklearn.metrics import explained_variance_score
```

```python
df = pd.read_csv("../my_work/clean_data.csv")
df = df.drop(df.columns[0], axis=1)
X = df.drop(["energy_consumption_per_annum"], axis=1)
y = df["energy_consumption_per_annum"]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Define the LightGBM model
lgb_model = lgb.LGBMRegressor()

# Train the model on the training data
lgb_model.fit(X_train, y_train) 
# Make predictions on the test data
y_pred = lgb_model.predict(X_test)

# Calculate the explained variance score
explained_variance = explained_variance_score(y_test, y_pred)
print("Explained variance:", explained_variance)
y_pred_train = lgb_model.predict(X_train)
explained_variance = explained_variance_score(y_train, y_pred_train)
print("Explained variance on train:", explained_variance)
```

```python
# Define the LightGBM model
lgb_model = lgb.LGBMRegressor()

# Define the grid of hyperparameters to search
param_grid = {'num_leaves': [30,40], #should be less than the number of features to avoid overfitting
              'max_depth': [-1, 5, 10],
              'min_data_in_leaf': [50, 100],
              'feature_fraction': [0.5, 0.8],
              'bagging_fraction': [0.5, 0.8],
              'learning_rate': [0.1, 0.01, 0.001],
              'n_estimators': [100, 500]
             }

# Define the grid search
grid_search = GridSearchCV(lgb_model, param_grid, cv=5, scoring='explained_variance')

# Fit the grid search to the data
grid_search.fit(X, y)

# Print the best hyperparameters
print("Best hyperparameters:", grid_search.best_params_)

# Make predictions on the test data
y_pred = grid_search.predict(X_test)

# Calculate the explained variance score
explained_variance = explained_variance_score(y_test, y_pred)
print("Explained variance:", explained_variance)
```
